```python
# Set the Seaborn style
sns.set(style="whitegrid")

# Create a subplot
f, ax = plt.subplots(figsize=(12, 8))

# Create a bar plot for Murder/Lakh
ax = sns.barplot(x='Murder/Lakh', y='STATE', data=data, color='skyblue', label='Murder/Lakh')

# Add labels and title
ax.set(xlabel='Murder per Lakh Population', ylabel='State', title='Murder Rates by State')

# Show the plot
plt.show()
```

```python
# Add Literacy
data['Literacy'] = [0.8663, 0.6702, 0.6538, 0.7219, 0.618, 0.8605, 0.7028, 0.7634, 0.8621, 0.887,
                    0.7803, 0.7555, 0.828, 0.6716, 0.6641, 0.7536, 0.94, 0.77, 0.9185, 0.6932,
                    0.8234, 0.7694, 0.7443, 0.9133, 0.7955, 0.7287, 0.8585, 0.7584, 0.6611, 0.8142,
                    0.8009, 0.728, 0.8722, 0.6768, 0.7882, 0.7636]

# Display the updated DataFrame
data
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
# Calculate Literate Population
data['Literate Population'] = data['Literacy'] * data['Population']

# Scatter plot
plt.scatter(data['IPC Crimes'], data['Literate Population'])
plt.title("IPC Crimes vs Literate Population")
plt.ylabel("Literate Population")
plt.xlabel("IPC Crimes")

# Show the plot
plt.show()
```

```python
# Set the Seaborn style
sns.set(style="whitegrid")

# Create a count plot for IPC Crimes by State
f, ax = plt.subplots(1, 1, figsize=(10, 6))
ax = sns.barplot(x='STATE', y='IPC Crimes', data=data, palette='rainbow')

# Rotate x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

# Set labels and title
ax.set(xlabel='State', ylabel='IPC Crimes', title='IPC Crimes by State')

# Show the plot
plt.show()
```

```python
# Assuming you have the 'data' DataFrame
data_murders = data.groupby('STATE')['Murders'].sum().sort_values()

# Select the top 5 states with the least number of murders
top_5_states_least_murders = data_murders.head(5)

# Set up the plot
f, ax = plt.subplots(figsize=(8, 6))

# Create a bar plot
sns.barplot(x=top_5_states_least_murders.index, y=top_5_states_least_murders, palette='coolwarm', ax=ax)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set title and labels
plt.title('Top 5 States with the Least Number of Murders')
plt.xlabel('State')
plt.ylabel('Number of Murders')

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Assuming you have the 'data' DataFrame
data_murders = data.groupby('STATE')['Murders'].sum().sort_values(ascending=False)

# Select the top 5 states with the maximum number of murders
top_5_states_max_murders = data_murders.head(5)

# Set up the plot
f, ax = plt.subplots(figsize=(8, 6))

# Create a bar plot
sns.barplot(x=top_5_states_max_murders.index, y=top_5_states_max_murders, palette='coolwarm', ax=ax)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Set title and labels
plt.title('Top 5 States with the Maximum Number of Murders')
plt.xlabel('State')
plt.ylabel('Number of Murders')

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Calculate literate population (product of literacy rate and population)
data['Literate Population'] = (data['Literacy'] / 100) * data['Population']

# Create a correlation matrix
correlation_matrix = data[['IPC Crimes', 'Literate Population']].corr()

# Set up the plot
plt.figure(figsize=(8, 6))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Set title
plt.title('Correlation Heatmap between IPC Crimes and Literate Population')

# Show the plot
plt.show()
```

```python
# Assuming 'data' is your DataFrame
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Extracting data
missing_persons = data['Missing Persons']
murders = data['Murders']  # Correct the column name if it's 'Muders'
literate_population = data['Literate Population']

# Creating the 3D scatter plot
ax.scatter(missing_persons, murders, literate_population, c='blue', marker='o')

# Adding labels
ax.set_xlabel('Missing Persons')
ax.set_ylabel('Murders')
ax.set_zlabel('Literate Population')

# Adding a title
plt.title('3D Scatter Plot of Missing Persons, Murders, and Literate Population')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame

# Select the columns for correlation
corr_columns = ["Murders", "IPC Crimes", "Cyber Crimes", "Missing Persons",
                "Sanctioned Police Force", "Murder/Lakh", "Crime/1000", "Population"]

# Calculate the correlation matrix
corrMatrix = data[corr_columns].corr()

# Set the style and size
sns.set(font_scale=1.10)
plt.figure(figsize=(10, 10))

# Create a heatmap with correlation values annotated
sns.heatmap(corrMatrix, vmax=.8, linewidths=0.01,
            square=True, annot=True, cmap='viridis', linecolor="white")

# Set the title
plt.title('Correlation between features')

# Show the plot
plt.show()
```
